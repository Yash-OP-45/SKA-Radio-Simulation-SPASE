```python

import matplotlib.pyplot as plt
import pandas as pd
data = pd.read_csv('ra vs time.csv')
```

```python
time = data['time'].apply(lambda x: int(x[0:2]))
ra_cal = data['ra_cal']
ra_NASA = data['ra_NASA']
```

```python
plt.figure(figsize=(20,10))
plt.subplot(1, 2, 1)
plt.scatter(time, ra_cal, label='RA (Calculated)')
plt.scatter(time, ra_NASA, label='RA (NASA)')
plt.xlabel('time(hr)')
plt.ylabel('RA (Degrees)')
plt.title('RA (Calculated) and RA (NASA) V/S Time')
plt.legend()
```
