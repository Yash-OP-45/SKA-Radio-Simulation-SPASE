```python

import matplotlib.pyplot as plt
import pandas as pd
data = pd.read_csv('dec vs time.csv')
```

```python
time = data['time'].apply(lambda x: int(x[0:2]))
dec_cal = data['dec_cal']
dec_NASA = data['dec_NASA']
```

```python
plt.figure(figsize=(20,10))
plt.subplot(1, 2, 1)
plt.scatter(time, dec_cal, label='Dec (Calculated)')
plt.scatter(time, dec_NASA, label='Dec (NASA)')
plt.xlabel('time(hr)')
plt.ylabel('Dec (Degrees)')
plt.title('Dec (Calculated) and Dec (NASA) V/S Time')
plt.legend()
```
