```python

import matplotlib.pyplot as plt
import pandas as pd
data = pd.read_csv('conversions.csv')
```

```python
time = data['time'].apply(lambda x: int(x[0:2]))
alt = data['alt']
az = data['az']
ra = data['ra']
dec = data['dec']
```

```python
plt.figure(figsize=(20,10))
plt.subplot(1, 2, 1)
plt.scatter(time, alt, label='Altitude')
plt.scatter(time, az, label='Azimuth')
plt.xlabel('time(hr)')
plt.ylabel('Alt/Az (Degrees)')
plt.axvline(0, color='black', linestyle='dashed')
plt.axhline(0, color='black', linestyle='dashed')
plt.title('Altitude and Azimuth V/S Time')
plt.legend()
```

```python
plt.figure(figsize=(20,10))
plt.subplot(1, 2, 2)
plt.scatter(time, ra, label='Right Ascension')
plt.scatter(time, dec, label='Declination')
plt.xlabel('time(hr)')
plt.ylabel('RA/Dec (Degrees)')
plt.title('Right Ascension and Declination V/S Time')
plt.legend()
```

```python
plt.figure(figsize=(20,10))
plt.subplot(1,2,2)
plt.plot(time,ra,label='Right Ascension')
plt.plot(time,dec,label="Declination")
plt.xlabel('time(hr)')
plt.ylabel('RA/Dec (Degree)')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(20,10))
plt.subplot(1, 2, 1)
plt.scatter(time, alt, label='Altitude')
plt.scatter(time, az, label='Azimuth')
plt.scatter(time, ra, label='Right Ascension')
plt.scatter(time, dec, label='Declination')
plt.xlabel('time(hr)')
plt.ylabel('RA/Dec (Degrees)')
plt.title('Combined Graph V/S Time')
plt.legend()
plt.show()
```
